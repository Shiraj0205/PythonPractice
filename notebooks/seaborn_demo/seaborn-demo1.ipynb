## seaborn

- Seaborn is a library mostly used for statistical plotting in Python. 
- It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

- Iris is a flowering plant, the researchers have measured various features of the different iris flowers and recorded them digitally.
- Iris dataset contains five columns such as Petal Length, Petal Width, Sepal Length, Sepal Width and Species Type.

```python
data = sns.load_dataset("iris")
data
sns.lineplot(data=data, x="sepal_length", y="sepal_width")
```

## Using matplotlib and seaborn together

```python
# Load data set
data = sns.load_dataset("iris")

sns.lineplot(data=data, x="sepal_length", y="sepal_width")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

# Set dark theme
sns.set_style("dark")

# Remove axis tick marks
sns.despine()

plt.show()
```
