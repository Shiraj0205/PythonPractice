# matplotlib

Matplotlib is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in a variety of formats. Developed by John D. Hunter in 2003, it enables users to graphically represent data, facilitating easier analysis and understanding. If you want to convert your boring data into interactive plots and graphs, Matplotlib is the tool for you.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# plt.scatter()
- Each point plotted individually â€” no implicit line between points.
- Handles large datasets better for scatter-type visualizations.

```python
x_coords = np.random.randint(-100, 100, (1, 5))
y_coords = np.random.randint(-100, 100, (1, 5))

plt.title('Random Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.scatter(x_coords, y_coords, color='blue', marker='o')
```

# plt.plot()
- General-purpose plotting (line plots by default).
- Connects points with a line by default.
- Better suited when you want both markers and connecting lines.


```python
x_coords = np.random.randint(0, 100, (1, 5))
y_coords = np.random.randint(0, 100, (1, 5))

x_coords = x_coords.flatten()
y_coords = y_coords.flatten()

plt.title('Random Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x_coords, y_coords, 'r*-', label='Line with Markers')
plt.legend()
```
