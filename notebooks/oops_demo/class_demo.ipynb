A class is a blueprint for creating objects (instances).

```python
class dog():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        return f"Name: {self.name} ,Age: {self.age}"
```

```python
obj = dog("Tommy", 5)   
print(obj.display_info())
```

## Instance methods â†’ operate on self (object data).
- Most common type of method in Python classes. They operate on a specific instance (object) of a class.
- The first parameter of an instance method is conventionally named self, which refers to the instance itself.
- Instance methods can access and modify both instance-specific attributes (data associated with a particular object)
- Use instance methods when the method's logic depends on or modifies the state of a particular object.

## Class methods â†’ operate on cls (class-level).
- They are often used for factory methods (creating instances with specific configurations) or for methods that modify class-level attributes.
- Class methods are defined using the @classmethod decorator above their definition.
- The first parameter of a class method is conventionally named cls, which refers to the class itself.
- Class methods can access and modify class-level attributes, but they cannot directly access instance-specific attributes. 


# Key Differences Summarized:





```python
class PasswordPolicy:

    min_length = 8
    require_special = True

    def __init__(self, require_special=True):
        self.min_length = min_length
        self.require_special = require_special

    @classmethod
    def is_valid(class, password):
        if len(password) < class.min_length
            return False
        return True
```
